```python
from bs4 import BeautifulSoup
```

```python
import requests
import pandas as pd
import urllib.parse
```

```python
website = 'https://www.cinemablend.com/reviews/'
```

```python
response = requests.get(website)
```

```python
#status code
response.status_code
```

```python
#Soup object
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
#Results
results = soup.find_all('div',{'class':'content'})
```

```python
#for loop
title = []
author = []
rating_count = []
review = []



for result in results:

        #title
        try:
            title.append(result.find('h3').get_text())
        except:
            title.append('NOT AVAILABLE')

        #author
        try:
            author.append(result.find('p',{'class':'byline'}).get_text())
        except:
           author.append('NOT AVAILABLE')

        #rating_count
        try:
            rating_count.append(result.find('span',{'class':'chunk rating'}).get_text())
        except:
            rating_count.append('NOT AVAILABLE')

        #review
        try:
            review.append(result.find('p', {'class':'synopsis'}).get_text())
        except:
            review.append('NOT AVAILABLE')
```

```python
movie_review = pd.DataFrame({'title':title,'author':author,'rating_count':rating_count,'review':review})
```

```python
movie_review
```

```python
movie_review['review']=movie_review['review'].apply(lambda x:x.strip('\n'))
```

```python
movie_review
```

```python
#for loop
title = []
author = []
rating_count = []
review = []


for i in range(1,11):
    website = 'https://www.cinemablend.com/reviews/page/'+ str(i)

    response = requests.get(website)

    soup = BeautifulSoup(response.content, 'html.parser')

    results = soup.find_all('div',{'class':'content'})

    for result in results:

        #title
        try:
            title.append(result.find('h3').get_text())
        except:
            title.append('NOT AVAILABLE')

        #author
        try:
            author.append(result.find('p',{'class':'byline'}).get_text())
        except:
           author.append('NOT AVAILABLE')

        #rating_count
        try:
            rating_count.append(result.find('span',{'class':'chunk rating'}).get_text())
        except:
            rating_count.append('NOT AVAILABLE')

        #review
        try:
            review.append(result.find('p', {'class':'synopsis'}).get_text())
        except:
            review.append('NOT AVAILABLE')


```

```python
movie_review = pd.DataFrame({'title':title,'author':author,'rating_count':rating_count,'review':review})
```

```python
movie_review
```

```python
movie_review['review']=movie_review['review'].apply(lambda x:x.strip('\n'))
```
